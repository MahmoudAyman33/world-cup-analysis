## 1. Team Performance Analysis:

Team: This is the team name, representing the team whose performance is being measured.

Runs: Total runs scored by the team.

Wickets: Total wickets taken by the team.

Runs_per_ball: Average runs scored per ball by the team.

Wicketball_prob: Probability of taking a wicket on each ball.

Econ: Economy rate (runs conceded per over) of the team.

4s: Total boundaries (fours) scored by the team.

6s: Total sixes scored by the team.

SR: Strike rate (balls per wicket) of the team.

Not_Out: Number of times a batsman remained not out for the team.

Mins: Total minutes spent batting or bowling.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#i will create a df called df1 that has the values of 'team' and 'runs'
runs_count = df['runs'].sum()
team_runs = df.groupby(['team'])['runs'].sum()
df1= pd.DataFrame(team_runs)
display(df1)
```

### Total Runs Scored by Each Team

```python
plt.figure(figsize=(5, 3))
sns.barplot(x=df1.index,y=df1['runs'], palette='viridis')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.title('Total Runs Scored by Each Team')
plt.xticks(rotation=45)
plt.show()
```

### Each Team with thier wkts

```python
wkts_count = df['wkts'].sum()
team_wkts = df.groupby(['team'])['wkts'].sum()
df2 = pd.DataFrame(team_wkts)
sns.barplot(data=df2, x=df2.index, y=df2['wkts'])
plt.title('Each Team with thier wkts')
```

### Each Team with thier bb_bf

```python
#runs_count = df['runs'].sum()
team_bb_bf = df.groupby(['team'])['bb_bf'].sum()
df3 = pd.DataFrame(team_bb_bf)

sns.barplot(data=df3, x=df3.index, y=df3['bb_bf'])
plt.title('Each Team with thier bb_bf')
```

#### for loop to show each team with its 4s, 6s

```python
teams = df['team'].unique()  
for team in teams:
    the_team = df[df['team'] == team]  
    total_4s = the_team['4s'].sum()  
    total_6s = the_team['6s'].sum()  

    plt.figure(figsize=(3,3))
    plt.pie([total_4s, total_6s], labels=['4s', '6s'], autopct='%1.1f%%')
    plt.axis('equal')
    plt.title(f'{team}: 4s vs 6s')
    plt.show()
```

### total Minutes for Each Team

```python
team_mins = df.groupby(['team'])['mins'].sum()
df4 = pd.DataFrame(team_mins)

sns.barplot(x = df4.index , y = df4['mins'])
plt.title("total Minutes for Each Team")
```

## Opposition and Ground Analysis:

team: The team name to identify which team is playing.

player: The player name to identify individual player performances.

opposition: The name of the opposition team against which the match is played. This column will allow you to analyze performance against different opponents.

bat_or_bowl: Indicates whether the player is batting or bowling. This will help in segregating batting and bowling performances.

runs: Total runs scored by the player if batting.

wkts: Total wickets taken by the player if bowling.

4s: Total number of 4s hit by the player if batting.

6s: Total number of 6s hit by the player if batting.

econ: Economy rate of the player if bowling.

sr: Strike rate of the player if batting.

not_out: Indicates whether the player was not out or not if batting.

### dropping the columns thats not related to the opposition and ground

```python
teams_players=df.drop(columns = ['bb_bf','wicketball_prob','runs_per_ball','ground','start_date','overs','mdns','inns','mins'])
teams_players = pd.DataFrame(teams_players)
teams_players
```

### splitting the data into  bawllers and  batters 

```python
bowl =  teams_players[teams_players['bat_or_bowl']=='bowl']
bowl.drop(columns = ['4s','6s','sr','not_out'],inplace = True)
bowl
```

```python
bat =  teams_players[teams_players['bat_or_bowl']=='bat']
bat.drop(columns = ['wkts','econ'],inplace= True)
bat
```

### the best teams in bowlling

```python
bowl
```

```python
best_teams_and_players = []

for team_name, player_name,opposition_name in zip(bowl['team'], bowl['player'],bowl['opposition']):
    the_team = bowl[bowl['team'] == team_name]
    the_player = the_team[the_team['player'] == player_name]
    the_opposition = bowl[bowl['opposition'] == opposition_name]

    best_runs_for_team = the_team['runs'].max()
    best_wkts_for_team = the_team['wkts'].max()
    best_runs_for_player = the_player['runs'].max()
    best_wkts_for_player = the_player['wkts'].max()
    
    best_teams_and_players.append((team_name, player_name, best_runs_for_team, best_wkts_for_team, best_runs_for_player, best_wkts_for_player,opposition_name))

best_teams_and_players = pd.DataFrame(best_teams_and_players, columns=['team', 'player','best_runs_for_team', 'best_wkts_for_team', 'best_runs_for_player', 'best_wkts_for_player','opposition_name'])
best_teams_and_players
```

## Temporal Analysis

```python
matches_by_date = df.groupby('start_date')
```

```python
average_runs_per_match = matches_by_date['runs'].mean()
```

```python
plt.figure(figsize=(20, 6))

plt.plot(average_runs_per_match.index, average_runs_per_match.values, marker='o', color='orange')

plt.title('Average Runs Scored per Match Over Time')
plt.xlabel('Match Date')
plt.ylabel('Average Runs per Match')
plt.xticks(rotation=70)
plt.grid(True)
plt.show()
```
